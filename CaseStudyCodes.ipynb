```python
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np
```

```python
contingency = pd.crosstab(df3['Furthest Recruiting Stage Reached'], df3['Application Source'])
stat, p, dof, expected = chi2_contingency(contingency)

alpha = 0.05

print('Significance=%.5g, p=%.5g' % (alpha, p))
if p <= alpha:
    print('-> Variables are associated')
else:
    print('-> Variables are not associated')
```
